# Soal 1: Pemahaman Teori
Jelaskan perbedaan antara Univariate, Bivariate dan Multivariate Analysis 

Jawaban:

<h2>Univariate</h2>

> merupakan metode analisis yang melibatkan satu (uni) variabel(variate) dengan mendefiniskan, me-summarize-kan, dan menganalisis pola dari satu variabel. 

Types: Frequency table, Bar Charts, Historgram, Pie Chart, Frequency Polygon

<h2>Bivariate</h2>

> merupakan metode analisis yang melibatkan dua (bi) variable (variate) dengan mencari insight dan relasi dari dua variabel.

Types: Scatter plot, Linear Correlation, Chi-square, Z-test, T-test

<h2>Multivariate</h2>

> merupakan metode analisis yang melibatkan lebih dari dua (multi) variable (variate) dalam satu waktu atau bersamaan.

Types: Cluster Analysis, Factor Analysis, Multiple Regression Analysis, Principal Component Analysis, Correspondence Analysis

# Soal 2. Persiapan Data

*   Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()

*   Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()



```python
#code here
dftwo = df.dropna()
dftwo = dftwo.groupby(by='species').mean()
dftwo
```

# Soal 3. Univariate Analysis
Buatlah sebuah analisa univariate dengan menggunakan perintah seperti berikut:

*   Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm seperti expected output berikut:
*   Buatlah sebuah plot histogram dari dataset awal (sebelum grouping) untuk kolom body_mass_g
*   Jelaskan insight yang diperoleh dari ke dua grafik tersebut.





```python
#code here
# Create canvas
plt.figure(figsize=(12, 8))

# Modify df into a percentage number in string type
# @param    :   df with type int or float 
# @return   :   a string with '%' character
def number(df):
    return "{:.1f}%".format(df)

# Plotting data into pie plot and labels = index with percentage of each data
plt.pie(dftwo['flipper_length_mm'], labels=dftwo.index, autopct=lambda var: number(var))

plt.show
```

```python
# Create canvas
plt.figure(figsize=(12, 8))

# Plotting data into histogram plot
plt.hist(df['body_mass_g'])

# Change xlabel='Body_mass_g' and ylabel='Count'
plt.xlabel('Body_mass_g')
plt.ylabel('Count')

plt.show
```

# Soal 4. Bivariate Analysis



*   Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar' seperti expected output di bawah
*   Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm' seperti expected dibawah
*   Jelaskan insight yang diperoleh dari hasil analisa bivariate tersebut






```python
#code here
# plotting dataframe
sns.catplot(x='sex', y='bill_length_mm', data=df, kind='bar')
```

```python
# set figure or canvas size
sns.set({'figure.figsize':(13,8)})

# set axes style into white
sns.set_style('white')

# plotting dataframe
sns.scatterplot(x='body_mass_g', y='flipper_length_mm', data=df, hue='bill_length_mm')
```
