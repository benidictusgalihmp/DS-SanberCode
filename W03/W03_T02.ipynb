Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(11.75,6.25))
ax = fig.add_subplot()

ax.plot(x, cos, label='Cosx', color='red')
ax.plot(x, cos3, label='3 Cosx', linestyle='--', color='green')
ax.plot(x, sin, label='Sin(x-90)', linestyle='dashdot', color='blue')
ax.plot(x, sin2, label='2 Sinx', marker='x', linestyle='None', color='#b2b41c')

ax.set_title('Grafik Cos Sin')
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')

plt.legend(loc='center')
plt.axis([-11.25, 11.25, -3.5, 3.5])
plt.show()
```
