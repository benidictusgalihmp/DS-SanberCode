```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
df_group = df.groupby(by='Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].agg('mean')
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
# code here
import numpy as np

# initialization
df_trans = df_group.T                       # transpose dataframe group
plt.figure(figsize=(18,8))                  # canvas
x = np.arange(df_trans.index.size)          # position of bar
width = 0.2                                 # width position of bar
arr_color = ['red', 'blue', 'purple', 'grey', 'yellow', 'green', 'pink', 'red', 'blue', 'purple', 'grey', 'yellow']

# update initial position of bar chart
for i in range(len(x)):
    x[i] = x[i] * 3

for i, j in enumerate(df_trans.columns):
    plt.bar(x + (i * width), df_trans[j], width, label= j, color=arr_color[i])

plt.xticks(x + 1.2, df_trans.index)
plt.title('Mean Sales Video Games By Genre')
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.legend(loc='upper right')

plt.show()
```

Jawab:

Pada region **NA_Sales** genre produk penjualan rata-rata **tertingginya** adalah **Platform** sedangkan **terendahnya** adalah **Adventure**.

Pada region **EU_Sales** genre produk penjualan rata-rata **tertingginya** adalah **Shooter** sedangkan **terendahnya** adalah **Adventure**.

Pada region **JP_Sales** genre produk penjualan rata-rata **tertingginya** adalah **Role-Playing** sedangkan **terendahnya** adalah **Shooter**.

Pada region **Other_Sales** genre produk penjualan rata-rata **tertingginya** adalah **Shooter** sedangkan **terendahnya** adalah **Adventure**.

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
# initialization
plt.figure(figsize=(18,8))                  # canvas
x = df_trans.index                          # position of bar
y_bottom = 0                                # initial bar of y-position
width = 0.7                                 # width of bar

for i, j in enumerate(df_trans.columns):
    plt.bar(x, df_trans[j], width, bottom= y_bottom, label= j, color=arr_color[i])
    y_bottom += df_trans[j]

plt.title('Mean Sales Video Games By Genre')
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.legend(loc='upper right')

plt.show()
```

Jawab: Region dengan total jumlah penjualan rata-rata **tertingginya** adalah **NA_Sales** sedangkan **terendahnya Other_Sales**.
