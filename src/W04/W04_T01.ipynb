# Soal 1. Menentukan jenis data categorical atau numeric
Jelaskan data kategorikal (kualitatif) dan data numerik (kuantitatif)



Jawab disini:

Data kategorikal merupakan data yang memiliki value kualitatif berupa kata, simbol, atau angka yang membawa beberapa fakta/opini data yang mendeskripsikan suatu kategori atau kelompok.

- Data nominal merupakan data yang tidak memiliki order dan digunakan untuk melabel suatu variabel tanpa value kuantitas. 

- Data ordinal merupakan data yang mirip seperti data nominal namun memiliki order.

Data numerik:

- Variabel diskrit merupakan data yang memiliki nilai terbatas.

- Variabel kontinu merupakan data yang tidak memiliki batas dan dapat berubah sewaktu-waktu kemudian nilainya dapat infinite ataupun finite dan tidak hanya bernilai sangat tinggi namun dapat juga bernilai sangat kecil dengan range fraction atau decimalnya sangat tinggi.

    - Data interval merupakan data yang memiliki jarak data atau interval yang dapat diukur dengan skala interval dapat dikalkulasikan dan titik zero dapat berubah - ubah.

    - Data rasio merupakan data yang mirip seperti interval namun titik zero bernilai 0 (absolut).

```python
# code here
# Set up the canvas
plt.figure(figsize=(12,8))

# size of the data marks (dot)
area = 20*df['WindAvgMPH']

# visualize data with scatterplot and cmap = coolwarm
plt.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], s=area, c=df['TempAvgF'], cmap='coolwarm')

# style title, xlabel, and ylabel
plt.title('Austin Weather')
plt.xlabel('Humidity Average (%)')
plt.ylabel('Dew Point Average (F)')
```

```python
# handling NaN value with fillna method ffill that replace value with
# last valid observation forward
df = df.fillna(method="ffill")

# set up the canvas
plt.figure(figsize=(12,8))

# size of the data marks (dot)
area = 20*df['WindAvgMPH']

# visualize data with scatterplot and cmap = coolwarm
plt.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], s=area, c=df['TempAvgF'], cmap='coolwarm')

# style title, xlabel, and ylabel
plt.title('Austin Weather')
plt.xlabel('Humidity Average (%)')
plt.ylabel('Dew Point Average (F)')
```

<h2>Soal 3. Visualisasi Data with Handling outliers </h2>

Pada tugas kali ini, kita akan melakukan handling outliers. Outliers ditemukan pada kolom 'House_Price' dengan menggunakan boxplot dari library seaborn seperti gambar berikut:

Dengan menggunakan Interquartile range, kita bisa temukan nilai data outliers dan meremove data outlier tersebut. 

*   Tentukan nilai batas atas dan batas bawah dari Interquartile range tersebut
*   Remove data outliers tersebut menggunakan batas atas dan batas bawah
*   Visualisasikan hasil data yang sudah d remove outliers seperti expected berikut:







**EXPECTED OUTPUT:**

Visualisasi boxplot:

![boxplot2](https://drive.google.com/uc?id=1qZc-lEa5FUzSQ_cyt4_GJQ0TZ8c6WjHX)





```python
#Code here
import seaborn as sns

# Calculate q1, q3 and interquartile
q1 = np.percentile(df2['House_Price'], 25)
q3 = np.percentile(df2['House_Price'], 75)
iqr = q3 - q1

# Calculate the lower and upper bound of data
lower_ = q1 - (iqr * 1.5)
upper_ = q3 + (iqr * 1.5)

# Print lower_ and upper_
print("Batas bawah & batas atas:")
print(lower_, upper_, end='\n')
print("Visualisasi boxplot:")

# Filter value data between lower and upper bound
dffilter = df2[(df2.House_Price < upper_) & (df2.House_Price > lower_)]

# Visualize the data with boxplot
sns.boxplot(x=dffilter.House_Price)
```
